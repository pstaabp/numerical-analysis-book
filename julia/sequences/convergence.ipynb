```julia
dfact(n::Int) = prod(1:2:n)
fact(n::Int) = prod(1:n)
```

```julia
xpts = Float64[8]
push!(xpts, xpts[1] + dfact(1)/(2^1 *fact(1)*8^1)*(37-36))
push!(xpts, xpts[2] - dfact(1)/(2^2 *fact(2)*8^3)*(37-36)^2)
push!(xpts, xpts[3] + dfact(3)/(2^3 *fact(3)*8^5)*(37-36)^2)
push!(xpts, xpts[4] - dfact(5)/(2^4 *fact(4)*8^7)*(37-36)^2)
push!(xpts, xpts[5] + dfact(7)/(2^5 *fact(5)*8^9)*(37-36)^2)
push!(xpts, xpts[6] - dfact(9)/(2^6 *fact(6)*8^11)*(37-36)^2)
```

```julia
eps = map(n -> abs(xpts[n] - sqrt(65)), 1:length(xpts))
```

```julia
log.(eps[3:end] ./ eps[2:end-1]) ./ log.(eps[2:end-1] ./ eps[1:end-2])
```

```julia
eps[end]/eps[end-1]
```
