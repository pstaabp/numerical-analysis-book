```julia
function exp_approx(x,n)
  sum = 0
  for j=0:n
    sum += x^j/factorial(j)
  end
  sum
end
```

```julia
fig = Figure()
ax = Axis(fig[1,1])
lines!(ax, -3..3, x->exp_approx(x,1))
lines!(ax, -3..3, x->exp_approx(x,2))
lines!(ax, -3..3, x->exp_approx(x,3))
lines!(ax, -3..3, x->exp_approx(x,5))
lines!(ax, -3..3, x->exp_approx(x,10))
lines!(ax, -3..3, exp)
fig
```
