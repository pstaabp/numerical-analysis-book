```julia
using ForwardDiff, CairoMakie
CairoMakie.activate!()
Makie.inline!(true)
```

```julia
f(x) = 3x*(x+1)*(x-2)^2
```

```julia
ForwardDiff.derivative(f,-1),ForwardDiff.derivative(f,0),ForwardDiff.derivative(f,2)
```

```julia
fig = Figure()
ax = Axis(fig[1,1])
lines!(ax, -1.5..3,x->3*(x+1)*x*(x-2)^2)
lines!(ax, -1.5..(-0.5),x->-27*(x+1))
lines!(ax, -0.5..0.5, x->12*x)
lines!(ax, 1.2..2.8, x->18*(x-2)^2)
fig
```
